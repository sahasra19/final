# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

def plot_KNN_error_rate(xdata, ydata):
    error_rate = []
    test_scores = []
    train_scores = []

    # Splitting the data into training and testing sets (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=1)

    for i in range(1, 40):  # Iterating over different values of k (from 1 to 39)
        # Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        # Fit the KNN model on the training set
        knn.fit(X_train, y_train)

        # Make predictions on the test set
        pred_i = knn.predict(X_test)

        # Calculate the error rate and append to the list
        error_rate.append(np.mean(pred_i != y_test))

        # Store the training and testing scores
        train_scores.append(knn.score(X_train, y_train))
        test_scores.append(knn.score(X_test, y_test))

    # Plot the Error Rate vs. K Value graph
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    # Find and display the maximum train score and corresponding K values
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print('Max train score: {:.2f}% at k = {}'.format(max_train_score * 100, list(map(lambda x: x + 1, train_scores_ind))))

    # Find and display the maximum test score and corresponding K values
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print('Max test score: {:.2f}% at k = {}'.format(max_test_score * 100, list(map(lambda x: x + 1, test_scores_ind))))

    return test_scores
```

```python
# @title
## TASK-7: Refer to MinMax Scaler provided in scikit-learn.
## Use MinMax scaling on the dataset, and see the performance of KNN on this minmax-scaled dataset.
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

def plot_KNN_with_scaling(xdata, ydata):
    # Step 1: Apply MinMax Scaling
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to a range [0, 1]

    # Step 2: Split the scaled data into train and test sets (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=1)

    error_rate = []
    test_scores = []
    train_scores = []

    # Step 3: Iterate over different values of k (from 1 to 39)
    for i in range(1, 40):
        # Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        # Fit the KNN model on the training set
        knn.fit(X_train, y_train)

        # Make predictions on the test set
        pred_i = knn.predict(X_test)

        # Calculate the error rate and append to the list
        error_rate.append(np.mean(pred_i != y_test))

        # Store the training and testing scores
        train_scores.append(knn.score(X_train, y_train))
        test_scores.append(knn.score(X_test, y_test))

    # Step 4: Plot the Error Rate vs. K Value graph
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value (After MinMax Scaling)')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    # Step 5: Find and display the maximum train and test scores
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print('Max train score (scaled): {:.2f}% at k = {}'.format(max_train_score * 100, list(map(lambda x: x + 1, train_scores_ind))))

    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print('Max test score (scaled): {:.2f}% at k = {}'.format(max_test_score * 100, list(map(lambda x: x + 1, test_scores_ind))))

    return test_scores


```

```python
## TASK-8:
## Plot the voronoi diagram for the TASK-7
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

def plot_voronoi_diagram(xdata, ydata, k_value=5):
    # Step 1: Apply MinMax Scaling to the data
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to a range [0, 1]

    # Step 2: Split the scaled data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=1)

    # Step 3: Train the KNN model with the given k_value
    knn = KNeighborsClassifier(n_neighbors=k_value)
    knn.fit(X_train, y_train)

    # Step 4: Create a meshgrid to plot decision boundaries
    h = 0.01  # Step size for the meshgrid
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Step 5: Plot the Voronoi diagram (decision boundaries)
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdYlBu)

    # Step 6: Plot the training points on top of the decision boundaries
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, edgecolors='k', marker='o', s=80, cmap=plt.cm.RdYlBu)
    plt.title(f'Voronoi Diagram for KNN (k={k_value})')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.colorbar()
    plt.show()

# Example usage
# plot_voronoi_diagram(xdata, ydata, k_value=5)
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
xdata = iris.data[:, :2]  # Take only the first two features for easier plotting
ydata = iris.target

# Call the function to plot the Voronoi diagram with k=5
plot_voronoi_diagram(xdata, ydata, k_value=5)

```

```python
## TASK-9: Use K-Fold cross validation on all the above classification experiments and present an analysis of the results you obtain.
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

def knn_cross_validation(xdata, ydata, k_value=5, n_splits=5):
    # Step 1: Apply MinMax Scaling to the data
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)  # Scale the features to a range [0, 1]

    # Step 2: Initialize the KNN classifier with the specified k_value
    knn = KNeighborsClassifier(n_neighbors=k_value)

    # Step 3: Initialize K-Fold Cross Validation
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=1)

    # Step 4: Apply K-Fold cross-validation
    scores = cross_val_score(knn, xdata_scaled, ydata, cv=kf, scoring='accuracy')

    # Step 5: Analyze the results
    print(f"K-Fold Cross Validation Results (k={k_value}, n_splits={n_splits}):")
    print(f"Accuracy for each fold: {scores}")
    print(f"Mean Accuracy: {scores.mean() * 100:.2f}%")
    print(f"Standard Deviation of Accuracy: {scores.std() * 100:.2f}%")

    # Step 6: Plot the accuracy for each fold
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, n_splits + 1), scores, marker='o', linestyle='-', color='b', label='Accuracy per Fold')
    plt.title(f'K-Fold Cross Validation Accuracy (k={k_value}, n_splits={n_splits})')
    plt.xlabel('Fold')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.show()

# Example usage
# knn_cross_validation(xdata, ydata, k_value=5, n_splits=5)
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
xdata = iris.data  # Features (sepal length, sepal width, etc.)
ydata = iris.target  # Target labels (species)

# Call the knn_cross_validation function to perform K-Fold cross-validation with k=5 and 5 splits
knn_cross_validation(xdata, ydata, k_value=5, n_splits=5)

```
